```
import os 
import time
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from azure.storage.blob import BlobServiceClient

# Set up Azure Storage credentials
storage_account_name = "uhgstoragee"
storage_account_key = "Dh5cvpN+yHtULbgFjHTj5gCw056yfq2QiNC9Kyvv8vfFFAidX37GjD/mjr9XA1r4axUMWIHqVbqa+AStQeS17Q=="  # Replace with your actual key
container_name = "uhgdata"

# Set up the Blob Service Client
blob_service_client = BlobServiceClient(account_url=f"https://{storage_account_name}.blob.core.windows.net", credential=storage_account_key)
container_client = blob_service_client.get_container_client(container_name)

# Set up the Chrome WebDriver
s = Service('/tmp/chrome/latest/chromedriver_linux64/chromedriver')  
options = Options()
options.binary_location = "/tmp/chrome/latest/chrome-linux/chrome"  
options.add_argument('headless')
options.add_argument('--disable-infobars')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument('--homedir=/tmp/chrome/chrome-user-data-dir')
options.add_argument('--user-data-dir=/tmp/chrome/chrome-user-data-dir')

driver = webdriver.Chrome(service=s, options=options)

# URL to scrape
url = "https://transparency-in-coverage.uhc.com/"

# Navigate to the URL
driver.get(url)

# Wait for the JavaScript to load
time.sleep(5)  # Adjust this as necessary for your specific page

# Get the page source and create a BeautifulSoup object
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find all file links (assuming they are in <a> tags)
items = soup.find_all('a')

# Initialize a list to store file URLs
file_urls = []
max_files = 1000

# Loop through the elements and extract file URLs
for item in items:
    if len(file_urls) >= max_files:
        break  # Stop when we reach 1000 files

    # Get the URL of the file (usually in href attribute)
    file_url = item.get('href', '')

    # Extract the filename from the URL
    file_name = os.path.basename(file_url.split('?')[0])  # Remove query parameters

    # Print the href and filename for debugging
    print(f"Link found: {file_url} | Filename: {file_name}")

    # Filter out non-file links and ensure the filename ends with .json
    if file_name.endswith('.json'):
        file_urls.append(file_url)

# Print the found file URLs
print(f"Found {len(file_urls)} files to download:")
for url in file_urls:
    print(url)  # Debugging: print each URL

# Download each file and directly upload to Azure Data Lake Storage
for file_url in file_urls:
    # Extract the filename from the URL
    file_name = os.path.basename(file_url.split('?')[0])  # Remove query parameters

    # Download the file content using requests in streaming mode
    print(f"Downloading and uploading {file_name}...")

    response = requests.get(file_url, stream=True)

    # Check if the request was successful
    if response.status_code == 200:
        # Get the blob client for the file
        blob_client = container_client.get_blob_client(file_name)

        # Upload the file to Azure Data Lake Storage in chunks
        blob_client.upload_blob(response.raw, overwrite=True)
        print(f"Uploaded {file_name} to ADLS Gen2 in container {container_name}")
    else:
        print(f"Failed to download {file_name}. Status code: {response.status_code}")

print(f"Total uploaded files: {len(file_urls)}")

# Close the Selenium driver
driver.quit()
```
